```python
import pandas as pd 
import numpy as np 
import time

from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

# Read csv
df = pd.read_csv('D:\EE551\\heart_2020_cleaned.csv')
```

```python
starttime=time.time()

# Data processing
for i in df.HeartDisease.values:
    if i  == 'No':
        df.HeartDisease.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.HeartDisease.replace(i, 1, inplace = True)
    
for i in df.Smoking.values:
    if i  == 'No':
        df.Smoking.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.Smoking.replace(i, 1, inplace = True)
        
for i in df.AlcoholDrinking.values:
    if i  == 'No':
        df.AlcoholDrinking.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.AlcoholDrinking.replace(i, 1, inplace = True)        

for i in df.Stroke.values:
    if i  == 'No':
        df.Stroke.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.Stroke.replace(i, 1, inplace = True)

for i in df.DiffWalking.values:
    if i  == 'No':
        df.DiffWalking.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.DiffWalking.replace(i, 1, inplace = True)        

for i in df.Sex.values:
    if i  == 'Female':
        df.Sex.replace(i, 0, inplace = True)
    elif i == 'Male':
        df.Sex.replace(i, 1, inplace = True)        

for i in df.AgeCategory.values:
    if i  == '18-24':
        df.AgeCategory.replace(i, 0, inplace = True)
    elif i == '25-29':
        df.AgeCategory.replace(i, 1/12, inplace = True)
    elif i == '30-34':
        df.AgeCategory.replace(i, 2/12, inplace = True)    
    elif i == '35-39':
        df.AgeCategory.replace(i, 3/12, inplace = True)
    elif i == '40-44':
        df.AgeCategory.replace(i, 4/12, inplace = True) 
    elif i == '45-49':
        df.AgeCategory.replace(i, 5/12, inplace = True)
    elif i == '50-54':
        df.AgeCategory.replace(i, 6/12, inplace = True)
    elif i == '55-59':
        df.AgeCategory.replace(i, 7/12, inplace = True)
    elif i == '60-64':
        df.AgeCategory.replace(i, 8/12, inplace = True) 
    elif i == '65-69':
        df.AgeCategory.replace(i, 9/12, inplace = True)
    elif i == '70-74':
        df.AgeCategory.replace(i, 10/12, inplace = True)
    elif i == '75-79':
        df.AgeCategory.replace(i, 11/12, inplace = True)
    elif i == '80 or older':
        df.AgeCategory.replace(i, 1, inplace = True)    
        
for i in df.PhysicalActivity.values:
    if i  == 'No':
        df.PhysicalActivity.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.PhysicalActivity.replace(i, 1, inplace = True)        

for i in df.Diabetic.values:
    if i  == 'No':
        df.Diabetic.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.Diabetic.replace(i, 1, inplace = True)
    elif i == 'No, borderline diabetes':
        df.Diabetic.replace(i, 1/3, inplace = True)    
    elif i == 'Yes (during pregnancy)':
        df.Diabetic.replace(i, 2/3, inplace = True)    

        
for i in df.GenHealth.values:
    if i  == 'Excellent':
        df.GenHealth.replace(i, 1, inplace = True)
    elif i == 'Very good':
        df.GenHealth.replace(i, 0.75, inplace = True)
    elif i == 'Good':
        df.GenHealth.replace(i, 0.5, inplace = True)    
    elif i == 'Fair':
        df.GenHealth.replace(i, 0.25, inplace = True)
    elif i == 'Poor':
        df.GenHealth.replace(i, 0, inplace = True)    
        
for i in df.Asthma.values:
    if i  == 'No':
        df.Asthma.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.Asthma.replace(i, 1, inplace = True) 
        
for i in df.KidneyDisease.values:
    if i  == 'No':
        df.KidneyDisease.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.KidneyDisease.replace(i, 1, inplace = True)      
        
for i in df.SkinCancer.values:
    if i  == 'No':
        df.SkinCancer.replace(i, 0, inplace = True)
    elif i == 'Yes':
        df.SkinCancer.replace(i, 1, inplace = True)      
        
for i in df.Race.values:
    if i  == 'White':
        df.Race.replace(i, 0, inplace = True)
    elif i == 'Hispanic':
        df.Race.replace(i, 1/5, inplace = True)
    elif i == 'Black':
        df.Race.replace(i, 2/5, inplace = True)    
    elif i == 'Asian':
        df.Race.replace(i, 3/5, inplace = True)
    elif i == 'American Indian/Alaskan Native':
        df.Race.replace(i, 4/5, inplace = True) 
    elif i == 'Other':
        df.Race.replace(i, 1, inplace = True)

endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')
```

```python
# Check the data type
df.info()
```

```python
# Data normalize
df = (df-df.min())/(df.max()-df.min())
df.describe()
```

```python
# Independent variable
X = df[['BMI', 'Smoking', 'AlcoholDrinking','Stroke'
        ,'PhysicalHealth','MentalHealth','DiffWalking','Sex'
        ,'AgeCategory','Diabetic','Race','PhysicalActivity','GenHealth'
        ,'SleepTime','Asthma','KidneyDisease','SkinCancer']].values
# Dependent variable
Y = df[ 'HeartDisease'].values
# Take a random 70% samples for training and the rest 30% for test
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB

starttime=time.time()
NB_Bernoulli=BernoulliNB()
NB_Bernoulli.fit(Xtrain,Ytrain)
NB_Bernoullipred=NB_Bernoulli.predict(Xtest)
# print(confusion_matrix(Ytest,NB_Bernoullipred ))
print("BernoulliNB Accuracy : ", accuracy_score(Ytest,NB_Bernoullipred))
endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')

starttime=time.time()
NB_Gaussian=GaussianNB()
NB_Gaussian.fit(Xtrain,Ytrain)
NB_Gaussianpred=NB_Gaussian.predict(Xtest)
# print(confusion_matrix(Ytest,NB_Gaussianpred))
print("\nGaussianNB Accuracy : ", accuracy_score(Ytest,NB_Gaussianpred ))
endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')

starttime=time.time()
NB_Multinomial=MultinomialNB()
NB_Multinomial.fit(Xtrain,Ytrain)
NB_Multinomialpred=NB_Multinomial.predict(Xtest)
# print(confusion_matrix(Ytest,NB_Multinomialpred))
print("\nMultinomialNB Accuracy : ", accuracy_score(NB_Multinomialpred, Ytest))
endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')
```

```python
from sklearn.linear_model import LogisticRegression

starttime=time.time()
LR = LogisticRegression(random_state = 0, max_iter=300)
LR.fit(Xtrain, Ytrain)
LRpred = LR.predict(Xtest)
print(confusion_matrix(Ytest, LRpred))
print ("Accuracy : ", accuracy_score(Ytest, LRpred))
endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')
```

```python
from sklearn.tree import DecisionTreeClassifier  

starttime=time.time()
DecisionTree=DecisionTreeClassifier(criterion='entropy', max_leaf_nodes=498)
DecisionTree.fit(Xtrain, Ytrain)
DecisionTreepred =DecisionTree.predict(Xtest)
print(confusion_matrix(Ytest, DecisionTreepred))
print(accuracy_score(Ytest, DecisionTreepred))
endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')
```

```python
from sklearn.ensemble import RandomForestClassifier

starttime=time.time()
RrandomForest=RandomForestClassifier()
RrandomForest.fit(Xtrain, Ytrain)
RrandomForestpred = RrandomForest.predict(Xtest)
print(confusion_matrix(Ytest, RrandomForestpred))
print(accuracy_score(Ytest, RrandomForestpred))
endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')
```

```python
from sklearn.neural_network import MLPClassifier

starttime=time.time()
ANN = MLPClassifier(hidden_layer_sizes=(30,30), 
                    activation='logistic', 
                    solver='adam',
                    learning_rate_init=0.001, 
                    random_state=0)
ANN.fit(Xtrain, Ytrain)
ANNpred = ANN.predict(Xtest)
print(confusion_matrix(Ytest, ANNpred))
print(accuracy_score(Ytest, ANNpred))
endtime=time.time()
runtime=endtime-starttime
print('Program running time:',runtime,'s')
```
